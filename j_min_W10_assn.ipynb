### 1.Create a .CSV file that includes all of the information shown above
To create a csv file, we need the csv module. We use with open function to open a desired csv file in the repository and use the mode w to create the file if it does not exist yet. Within the with open clause we use writer.writerow function to input value to each cell in this table. 

```python
# import csv module
import csv
# use with open function to write a csv file
with open('arrival_delay.csv', mode='w') as file:
    writer = csv.writer(file)
    writer.writerow(['Airlines','Arrival','Los Angeles','Phoenix','San Diego','San Francisco','Seattle'])
    writer.writerow(['ALASKA','on time','497','221','212','503','1841'])
    writer.writerow(['ALASKA','delayed','62','12','20','102','305'])
    writer.writerow(['AM WEST','on time','694','4840','383','320','201'])
    writer.writerow(['AM WEST','delayed','117','415','65','129','61'])
```

### 2.Read .CSV file from GitHub repository and combine and reshape data in Pandas to tidy and transform data
To read a csv file and combine and reshape data we need to use pandas. After reading the csv file we transfer the wide format to long format by using airlines and arrival as id variables and destinations as value variables.

```python
# import pandas
import pandas as pd
# use read_csv function to read the csv file
arrival_delay = pd.read_csv('arrival_delay.csv')
# use melt function to tidy and transform data
ad_long = pd.melt(arrival_delay, id_vars=['Airlines','Arrival'], value_vars=['Los Angeles','Phoenix','San Diego','San Francisco','Seattle'])
ad_long
```

### 3.Using reshaped/transformed data, perform analysis to compare the arrival delays for the two airlines
For each city, which airline had the best on time performance? To decide a best on time performance for each city, we can calculate the on time rate for each airline by using on time numbers divided by total numbers. Then for each city we compare the on time percentage of two airlines and print the bettwe one. 
Which airline had the best overall on time performance? To decide a best overall on time performance, we can calculate overall on time percentage for each airline by using overall on time numbers divided by total numebers. Then we compare the percentage and print the better one. 

```python
# import numpy
import numpy as np
# create a dataframe of on time and delayed numbers for each city
dest_list['destination'] = pd.DataFrame(['Los Angeles','Phoenix','San Diego','San Francisco','Seattle'])
alo = pd.DataFrame(list(ad_long['value'][(ad_long['Airlines']=='ALASKA') & (ad_long['Arrival']=='on time')]))
ald = pd.DataFrame(list(ad_long['value'][(ad_long['Airlines']=='ALASKA') & (ad_long['Arrival']=='delayed')]))
awo = pd.DataFrame(list(ad_long['value'][(ad_long['Airlines']=='AM WEST') & (ad_long['Arrival']=='on time')]))
awd = pd.DataFrame(list(ad_long['value'][(ad_long['Airlines']=='AM WEST') & (ad_long['Arrival']=='delayed')]))
# calculate the on time percentage
dest_list_al = np.array(alo/(alo+ald))
dest_list_aw = np.array(awo/(awo+awd))
# compare the on time percentage
dest_list['best airline'] = np.where(dest_list_al > dest_list_aw, 'ALASKA', 'AM WEST')
# print out the result
print(dest_list[['destination','best airline']])
```

```python
# calculate the overall on time and delayed numbers for each airline
alo = ad_long['value'][(ad_long['Airlines']=='ALASKA') & (ad_long['Arrival']=='on time')].sum()
ald = ad_long['value'][(ad_long['Airlines']=='ALASKA') & (ad_long['Arrival']=='delayed')].sum()
awo = ad_long['value'][(ad_long['Airlines']=='AM WEST') & (ad_long['Arrival']=='on time')].sum()
awd = ad_long['value'][(ad_long['Airlines']=='AM WEST') & (ad_long['Arrival']=='delayed')].sum()
# compare the percentages and print out the result
if alo/(alo+ald) > awo/(awo+awd):
    print('ALASKA had the best overall on time performance')
else:
    print('AM WEST had the best overall on time performance')
```
