```python
# remove first three rows
a = a[3:]
# use len function to get the length of a
length = len(a)
# use np.arange to arange a, start at 2,step 3, end at 3
row1_num = np.arange(1,length,3)
step = 0
# create a empty list to store first arange list
row1 = []
#create a for loop, if variable step in row1_num, store that row to row1[]
for i in a:
    if(step in row1_num):
        row1.append(i)
    step = step+1
#print row1 
row1
```
